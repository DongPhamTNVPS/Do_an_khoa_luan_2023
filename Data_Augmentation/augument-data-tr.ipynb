```python
from keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array, array_to_img

# Khởi tạo ImageDataGenerator với các tham số tùy chỉnh
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')

# Load ảnh từ thư mục
img = load_img('/kaggle/input/data-benh-da-lieu/files/redness/29/front.jpg')

# Chuyển đổi ảnh thành mảng numpy
x = img_to_array(img)

# Thêm một chiều mới cho mảng numpy
x = x.reshape((1,) + x.shape)

# Tạo ra 10 ảnh mới từ ảnh gốc
i = 0
output_file = '/kaggle/working/output/data-benh-da-lieu/files/redness/29'
for batch in datagen.flow_from_directory(
    '/kaggle/input/data-benh-da-lieu/files/acne',
    target_size=(256, 256),
    batch_size=1,
    save_to_dir=output_file,
    save_prefix='aug_',
    save_format='jpeg'):
    i += 1
    if i >= 10:
        break

# Hiển thị thông báo khi hoàn thành
print(f'{i} ảnh đã được tạo và ghi vào thư mục {output_file}!')
```
